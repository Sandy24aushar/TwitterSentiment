```python
data = pd.read_csv("/content/training.1600000.processed.noemoticon.csv", encoding="latin-1", header=None)
```

```python
data.head()
```

```python
data.rename(columns={0:"target", 1:"ids", 2:"date", 3:"flag", 4:"user", 5:"text"}, inplace=True)
```

```python
data
```

```python
red_data=data.drop(labels=data.columns[1:5],axis=1,inplace=False)
```

```python
red_data
```

```python
def normalize_tweet(tweet_text):
  return tweet_text.lower()
```

```python
with ThreadPoolExecutor(max_workers=4) as pool:
  red_data["text"]=list(pool.map(normalize_tweet, list(red_data["text"])))
```

```python
red_data
```

```python
regex_pattern = r'^@[a-zA-z0-9]|^#[a-zA-Z0-9]|\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*|\W+|\d+|<("[^"]*"|\'[^\']*\'|[^\'">])*>|_+|[^\u0000-\u007f]+'
```

```python
def rem_noi_tok(tweet_text):
  return re.sub(pattern=regex_pattern, repl=' ', string=tweet_text)
```

```python
with ThreadPoolExecutor(max_workers=4) as pool:
  red_data["text"] = list(pool.map(rem_noi_tok, list(red_data["text"])))
```

```python
red_data
```

```python
def tokz_tweet_text(tweet_text):
  return word_tokenize(tweet_text)
```

```python
with ThreadPoolExecutor(max_workers=4) as pool:
  red_data["text"] = list(pool.map(tokz_tweet_text, list(red_data["text"])))
```
